# Task 1 - Data Collection

In this initial phase of the project, our primary goal is to securely fetch the raw data required for our machine learning pipeline. 

## Objectives:
1. **Authentication**: Set up the environment with the Kaggle API.
2. **Data Downloading**: Securely download the historical housing records from Ames, Iowa.
3. **Data Storage**: Extract and store the raw dataset in our local `inputs/datasets/raw` directory for future processing.

The dataset is sourced from **Kaggle** and contains various features of residential homes along with their final sale prices, which will serve as the target variable for our predictive model.

---
## 1. Environment Setup

In this section, we will configure the working directory and set up the Kaggle API authentication. This is required to securely fetch the dataset from the Kaggle servers.

```python
import os

# Change the current working directory to the project's root
os.chdir("..")

# Set the current directory as the Kaggle configuration directory
os.environ['KAGGLE_CONFIG_DIR'] = os.getcwd()

# Set read and write permissions for the owner of the kaggle.json file
!chmod 600 kaggle.json
```

---
## 2. Fetching Data from Kaggle

Now that the environment is authenticated, we will define the specific Kaggle dataset path and execute the download directly into our workspace.

```python
KaggleDatasetPath = "codeinstitute/housing-prices-data"
DestinationFolder = "inputs/datasets/raw"

# Ensure the destination folder exists before downloading the data
os.makedirs(DestinationFolder, exist_ok=True)

# Download the dataset from Kaggle to the destination folder
!kaggle datasets download -d {KaggleDatasetPath} -p {DestinationFolder}
```

---
## 3. Data Storage

Finally, we will extract the downloaded dataset and store the raw `.csv` file inside our local `inputs/datasets/raw` directory. We will also remove the downloaded `.zip` file to keep our workspace clean.

```python
# Unzip the downloaded file and remove the .zip file afterwards
!unzip {DestinationFolder}/*.zip -d {DestinationFolder} \
  && rm {DestinationFolder}/*.zip

print("Dataset downloaded and unzipped successfully!")
```
